```python
# 划分数据集
import numpy as np
import scipy.io as sio
import pandas as pd

# 设置小区
districts = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# 加载 district_user_info.mat 文件
district_user_info = sio.loadmat('district_user_info.mat')

user_id = 0
cluster = 1
weekload = 2

# 创建字典来存储每个小区的数据集
data_dict = {}

for district in districts:
    if district in ['A', 'B', 'C', 'D', 'E', 'F', 'G']:
        district_data = district_user_info[f'{district}_user_data']
        user_ids = np.array(district_data[:, user_id], dtype=int)
        load = np.array([user[weekload] for user in district_data])

        history_days = 150
        F_existing_days = 2
        F_future_days = 19

        # 历史数据
        history_data = load[:, :, :history_days]

        # F小区已有数据
        F_existing_data = load[:, :, history_days:history_days + F_existing_days]

        # F小区未来数据
        F_future_data = load[:, :, history_days + F_existing_days:]

        # 划分训练集、验证集、测试集
        train_ratio = 0.8
        validation_ratio = 0.1
        test_ratio = 0.1

        train_data = history_data[:, :, :int(history_days * train_ratio)]
        validation_data = history_data[:, :, int(history_days * train_ratio):int(history_days * (train_ratio + validation_ratio))]
        test_data = history_data[:, :, int(history_days * (train_ratio + validation_ratio)):]

        print(f'{district}_user_data')
        print(f"训练集形状: {train_data.shape}")
        print(f"验证集形状: {validation_data.shape}")
        print(f"测试集形状: {test_data.shape}")
        print(f"F小区已有数据形状: {F_existing_data.shape}")
        print(f"F小区未来数据形状: {F_future_data.shape}")

        # 创建包含多维数组的 DataFrame
        df = pd.DataFrame({
            'user_id': user_ids,
            'weekload': [arr for arr in load],  # 将每个用户的 load 数据作为一个单独的二维数组存储
            'train_data': [arr for arr in train_data],
            'validation_data': [arr for arr in validation_data],
            'test_data': [arr for arr in test_data],
            'F_existing_data': [arr for arr in F_existing_data],
            'F_future_data': [arr for arr in F_future_data]
        })

        data_dict[f'{district}_user_data'] = df

# 保存处理后的数据
mat_data = {}
for district, df in data_dict.items():
    # 将 DataFrame 转换为 numpy 数组
    user_data = df.to_numpy()
    # 将 numpy 数组存储到 mat_data 字典中
    mat_data[district] = user_data

# 保存 mat 文件
sio.savemat('processed_data_dict.mat', mat_data)
print("处理后的数据已保存到 processed_data_dict.mat")
```

```python
# 将训练集，测试集和验证集按时间序列（单位：小时）排好 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, MinMaxScaler
import scipy.io as sio
import numpy as np

data = sio.loadmat('processed_data_dict.mat')
user_id = 0
load_all = 1
train_load = 2
val_load = 3
test_load = 4
exist_load = 5

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

train_time_series = {}
for district in data:
    if district.startswith('__'):
        continue
    district_data = data[district]
    l_data = np.array([user[train_load] for user in district_data])
    print(l_data.shape)
    num_users, num_hours, num_days = l_data.shape
    transformed_l_data = np.zeros((num_users, num_hours * num_days))
    for i in range(num_users):
         for j in range(num_days):
            for k in range(num_hours):
                transformed_l_data[i, j * num_hours + k] = l_data[i, k, j]
                train_time_series[district] = transformed_l_data    
    transformed_l_data = np.array(transformed_l_data)
    print(transformed_l_data.shape)     
    train_time_series[district] = transformed_l_data

```

```python
val_time_series = {}
for district in data:
    if district.startswith('__'):
        continue
    district_data = data[district]
    v_data = np.array([user[val_load] for user in district_data])
    print(v_data.shape)
    num_users, num_hours, num_days = v_data.shape
    transformed_v_data = np.zeros((num_users, num_hours * num_days))
    for i in range(num_users):
         for j in range(num_days):
            for k in range(num_hours):
                transformed_v_data[i, j * num_hours + k] = v_data[i, k, j]
                val_time_series[district] = transformed_v_data    
    transformed_v_data = np.array(transformed_v_data)
    print(transformed_v_data.shape)     
    val_time_series[district] = transformed_v_data
```

```python
test_time_series = {}
for district in data:
    if district.startswith('__'):
        continue
    district_data = data[district]
    t_data = np.array([user[test_load] for user in district_data])
    print(t_data.shape)
    num_users, num_hours, num_days = t_data.shape
    transformed_t_data = np.zeros((num_users, num_hours * num_days))
    for i in range(num_users):
         for j in range(num_days):
            for k in range(num_hours):
                transformed_t_data[i, j * num_hours + k] = t_data[i, k, j]
                test_time_series[district] = transformed_t_data    
    transformed_t_data = np.array(transformed_t_data)
    print(transformed_t_data.shape)     
    val_time_series[district] = transformed_t_data
```

```python
# 对用户数维度加和，归一化
scaler = MinMaxScaler(feature_range=(0, 1))
for district in train_time_series:
        train_time_series[district] = np.sum(train_time_series[district],axis=0)
        train_time_series[district] = scaler.fit_transform(train_time_series[district].reshape(-1,1))
        print(train_time_series[district].shape)
```

```python
for district in val_time_series:
        val_time_series[district] = np.sum(val_time_series[district],axis=0)
        val_time_series[district] = scaler.fit_transform(val_time_series[district].reshape(-1,1))
        print(val_time_series[district].shape)
```

```python
for district in test_time_series:
        test_time_series[district] = np.sum(test_time_series[district],axis=0)
        test_time_series[district] = scaler.fit_transform(test_time_series[district].reshape(-1,1))
        print(test_time_series[district].shape)
```

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Feb  3 16:31:10 2022

@author: ozancan ozdemir
"""

import matplotlib.pyplot as plt
import pandas as pd
import torch
import torch.nn as nn
from torch.autograd import Variable
from sklearn.preprocessing import MinMaxScaler
import torch.nn.functional as F

x_train = torch.tensor(x_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
x_val = torch.tensor(x_val, dtype=torch.float32)
y_val = torch.tensor(y_val, dtype=torch.float32)


class CNNLSTM(nn.Module):
    def __init__(self, input_size, output_size,hidden_size,num_layers):
        super(CNNLSTM, self).__init__()

        self.conv1 = nn.Conv1d(input_size, 64, kernel_size=1, stride=1)
        self.conv2 = nn.Conv1d(64,32,kernel_size=1, stride = 1, padding=1)
        self.batch1 =nn.BatchNorm1d(32)
        self.conv3 = nn.Conv1d(32,32,kernel_size=1, stride = 1, padding=1)
        self.batch2 =nn.BatchNorm1d(32)
        self.LSTM = nn.LSTM(input_size=5, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True)
        self.fc1 = nn.Linear(32*hidden_size, output_size)
        #self.fc2 = nn.Linear(1, 1)
        

    def forward(self, x):
        #in_size1 = x.size(0)  # one batch
        x = F.selu(self.conv1(x))
        x = self.conv2(x)
        x = F.selu(self.batch1(x))
        x = self.conv3(x)
        x = F.selu(self.batch2(x))
        x, h = self.LSTM(x) 
        x = torch.reshape(x,(x.shape[0],x.shape[1]*x.shape[2]))
        #in_size1 = x.size(0)  # one batch
        #x = x.view(in_size1, -1)
        # flatten the tensor x[:, -1, :]
        x = self.fc1(x)
        output = torch.sigmoid(x)
        #output = self.fc2(x)

    
        
        return output


model = CNNLSTM(24, 1,2, 3)
print(model)


num_epochs = 800
learning_rate = 0.01
criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
#optimizer = torch.optim.SGD(lstm.parameters(), lr=learning_rate)
loss_list = []
val_loss_list = []
# Train the model
for epoch in range(num_epochs):
    outputs = model(x_train)
    optimizer.zero_grad()
    
    # obtain the loss function
    loss = criterion(outputs, y_train)
    loss_list.append(loss)
    loss.backward()
    
    optimizer.step()
    if epoch % 100 == 0:
        model.eval()
        with torch.no_grad():
            val_outputs = model(x_val)
            val_loss = criterion(val_outputs, y_val)
            val_loss_list.append(val_loss.item())
            print(f"Epoch: {epoch}, Train Loss: {loss.item():.5f}, Val Loss: {val_loss.item():.5f}")
```
